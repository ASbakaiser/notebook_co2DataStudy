# Imports librairies

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

# import data from csv

```python
data = pd.read_csv("co2_emission_by_countries.csv", encoding="latin")
```

# reading data as dataframe

```python
df = pd.DataFrame(data)
```

```python
df
```

# Remove usless columns as "Calling code" and "Code"

```python
df1 = df.drop(columns=['Calling Code','Code'])
```

```python
df1
```

# Rename columns of select easier

```python
df1 = df1.rename(columns={'CO2 emission (Tons)': 'coTons', 'Population(2022)' : 'population'})
```

```python
df1
```

# Sorting the different columns to find correlations or anomalies

```python
df1 = df1.sort_values('population', ascending = False)
```

```python
df1
```

# as we can see, we have a lot of NaN value let's see in 2020 how many Country contain NaN information in this data set

# create filter for geting all data where year is 2020

```python
dfSelect2020 = df1[df1['Year'] == 2020]
```

```python
dfSelect2020
```

# create filter for return all rows where one value have NaN on all columns

```python
dfNull2020 = dfSelect2020[dfSelect2020.isna().any(axis=1)]
```

```python
dfNull2020['Country']
```

# how many are them ?

```python
len(dfNull2020.index)
```

# pure curiosity, let's compare the most polluting to this group of countries

```python
the31Union = dfNull2020['coTons'].sum()
theMostPoluating = dfSelect2020.nlargest(1, 'coTons')
```

```python
x_value = [theMostPoluating['Country'].values[0], 'U34']
y_value = [theMostPoluating['coTons'].values[0], the31Union]

plt.bar( x_value, y_value)

plt.show()
```

# I'm curious to find out the number of countries required to match the CO2 emissions of the United States

```python
dfSorted2020 = dfSelect2020.sort_values('coTons', ascending=True)
dfSorted2020
```

```python
i = 0
countriesSumsCo2 = 0
dfCountriesSums = pd.DataFrame(columns=df1.columns)
```

```python
while theMostPoluating['coTons'].values[0] > countriesSumsCo2 :
    
    countriesSumsCo2 += dfSorted2020['coTons'].values[i] 

    row = pd.DataFrame([dfSorted2020.iloc[i]])
    dfCountriesSums = pd.concat([dfCountriesSums, row])

    
    i += 1
print(str(i) + " " + str(countriesSumsCo2))
    
```

```python
dfCountriesSums
```

# After observing this, i understand that CO2 emissions are cumulative from 1700 rather than being based annualy starting from 1700.

# and when we see the united state at 417 billion tons co2Tons that confirm the data 'coTons' is cumulative and not based annualy

# let's get 10 country who generated the most of c02 in 2020 

```python
df2020Big = dfSelect2020.nlargest(10, 'coTons')
```

```python
df2020Big
```

# Getting data from all year for the 10 selected

```python
allYearCO2Big = df1[df1['Country'].isin(df2020Big['Country'])]
```

```python
allYearCO2Big
```

# creating pivot_table for setting year as index and country as column for Emission though year on graph

```python
pivot_table = allYearCO2Big.pivot_table(values='coTons', index='Year', columns='Country')
```

```python
pivot_table.plot()
plt.ylabel('CO2 TONS')
plt.title('CO2 Emissions over time')
plt.show()
```

# well let's now see from this ten coutry who's generated most co2 by Area we gonna take 2020 data for the first step

```python
df2020BigByArea = df2020Big
df2020BigByArea['coByArea'] = (df2020BigByArea['coTons'] / df2020BigByArea['Area'])
```

```python
df2020BigByArea
```

```python
df2020BigByArea.plot.bar(x='Country', y='coByArea', rot=30, figsize=(10,5))
plt.show()
```

# let's get coByArea from 1700 

```python
allYearCoAreaBig = allYearCO2Big.copy()
allYearCoAreaBig['coByArea'] = (allYearCoAreaBig['coTons'] / allYearCoAreaBig['Area'])
```

```python
allYearCoAreaBig
```

# create the pivot table for getting the graph

```python
pivot_table = allYearCoAreaBig.pivot_table(values='coByArea', index='Year', columns='Country')
```

```python
pivot_table.plot()
plt.ylabel('CO2 TONS')
plt.title('CO2 Emissions over time')
plt.show()
```

# i want to have a better vision with a dynamic graph, let's use plotly 

```python
fig = go.Figure()

for country in pivot_table.columns:
    fig.add_trace(
        go.Scatter(
            x=pivot_table.index,
            y=pivot_table[country],
            name=country
        )
    )

fig.update_layout(
    title="CO2 By States from 1700",
    xaxis_title="Years",
    yaxis_title="coByYear"
)

fig.show()
```
